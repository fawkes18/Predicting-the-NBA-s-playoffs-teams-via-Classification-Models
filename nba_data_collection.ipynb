```python
import requests
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
from pandas.io.html import read_html
from bs4 import Comment
from lxml import html
import warnings
warnings.filterwarnings('ignore')
```

##### 2019-20

```python
years = 2020

url = 'https://www.basketball-reference.com//leagues/NBA_{}.html'.format(years)

response = requests.get(url)

soup = BeautifulSoup(response.text, 'html.parser')

comments = soup.find_all(string=lambda text: isinstance(text, Comment))

tables = []
for each in comments:
    if 'table' in each:
        try:
            tables.append(pd.read_html(each)[0])
        except:
            continue
```

```python
misc20 = tables[6]
shoot20 = tables[7]
opshoot20 = tables[8]
```

```python
misc20.columns = misc20.columns.droplevel()
misc20 = misc20[['Team', 'W', 'Pace', 'FTr', '3PAr', 'TS%']]
```

```python
misc20.drop(misc20.index[30], inplace = True)
misc20.set_index('Team', inplace = True)
```

```python
misc20['W'] = misc20['W'] * 1.2
misc20['W'] = misc20['W'].round()
```

```python
shoot20.columns = shoot20.columns.droplevel()
shoot20.columns = shoot20.columns.droplevel()
```

```python
shoot20_1 = shoot20.iloc[:, 0:12]
shoot20_2 = shoot20.iloc[:, 12:19]
shoot20_3 = shoot20.iloc[:, 19:20]
shoot20_4 = shoot20.iloc[:, 21:22]
shoot20_5 = shoot20.iloc[:, 23:26]
```

```python
shoot20_1.rename(columns={'2P': '2p_fga','0-3': '0_3_fga', '3-10': '3_10_fga', '10-16': '10_16_fga', '16-3pt': '16_3pt_fga', '3P': '3p_fga'}, inplace = True)
shoot20_2.rename(columns={'2P': '2p_fg', '0-3': '0_3_fg', '3-10': '3_10_fg', '10-16': '10_16_fg', '16-3pt': '16_3pt_fg', '3P': '3pfg', "%Ast'd": '2pt_astd'}, inplace = True)
shoot20_3.rename(columns={'%FGA': 'dunks_fga'}, inplace = True)
shoot20_4.rename(columns={'%FGA': 'layup_fga'}, inplace = True)
shoot20_5.rename(columns={"%Ast'd": '3p_astd', '%3PA': 'corner_fga', '3P%': 'corner_fg'}, inplace = True)


all_shoot = [shoot20_2, shoot20_3, shoot20_4, shoot20_5]

shoot20 = shoot20_1.join(all_shoot, how = 'outer')

shoot20.drop(shoot20.index[30], inplace = True)
```

```python
opshoot20.columns = opshoot20.columns.droplevel()
opshoot20.columns = opshoot20.columns.droplevel()
```

```python
opshoot20_1 = opshoot20.iloc[:, 0:12]
opshoot20_2 = opshoot20.iloc[:, 12:19]
opshoot20_3 = opshoot20.iloc[:, 19:20]
opshoot20_4 = opshoot20.iloc[:, 21:22]
opshoot20_5 = opshoot20.iloc[:, 23:26]

opshoot20_1.rename(columns={'0-3': '0_3_fga', '3-10': '3_10_fga', '10-16': '10_16_fga', '16-3pt': '16_3pt_fga', '3P': '3p_fga', '2P': '2p_fga'}, inplace = True)
opshoot20_2.rename(columns={'0-3': '0_3_fg', '3-10': '3_10_fg', '10-16': '10_16_fg', '16-3pt': '16_3pt_fg', '3P': '3pfg', '2P': '2p_fg'}, inplace = True)
opshoot20_3.rename(columns={'%FGA': 'opp_dunks_fga'}, inplace = True)
opshoot20_4.rename(columns={'%FGA': 'opp_layup_fga'}, inplace = True)
opshoot20_5.rename(columns={"%Ast'd": 'opp_3p_astd', '%3PA': 'opp_corner_fga', '3P%': 'opp_corner_fg'}, inplace = True)

all_opshoot = [opshoot20_2, opshoot20_3, opshoot20_4, opshoot20_5]

opshoot20 = opshoot20_1.join(all_opshoot, how = 'outer')

opshoot20.set_index('Team', inplace = True)
opshoot20.drop(opshoot20.index[30], inplace = True)
```
